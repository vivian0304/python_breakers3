```python
## corona.xlsx 파일엔 1월20일 부터 1월30일까지의 중국과 타국에서의 COVID19 확진자 추이가 담겨있다.

import pandas as pd

# 이 엑셀을 pandas.DataFrame으로 불러온 뒤 '한국' column을 만들어서 다음의 데이터를 추가하고 covid19.csv로 저장하기
corona = pd.read_excel('C:\\Users\\82107\\Desktop\\5주차QUEST_데이터셋\\corona.xlsx', usecols="A, B, C", index_col=0) # 엑셀 파일 읽어오는 과정
corona["한국"] = [1, 1, 2, 3, 4, 7] # 한국 데이터 추가
corona.to_csv("C:\\Users\\82107\\Desktop\\covid19.csv", encoding="utf-8-sig") # covid19.csv로 저장

# covid19.csv를 불러와서 원본 데이터는 sheet1에, 중국 데이터는 sheet2에, 타국 데이터는 sheet3에, 한국 데이터는 sheet4에 저장하고 covid19_new.xlsx로 저장하기
covid19 = pd.read_csv('C:\\Users\\82107\\Desktop\\covid19.csv') # covid19.csv를 불러옴
covid19_1 = covid19[['날짜','중국','타국','한국']] # 원본 데이터
covid19_2 = covid19[['날짜','중국']] # 중국 데이터
covid19_3 = covid19[['날짜','타국']] # 타국 데이터
covid19_4 = covid19[['날짜','한국']] # 한국 데이터

with pd.ExcelWriter('C:\\Users\\82107\\Desktop\\covid19_new.xlsx', engine='xlsxwriter') as writer: # covid19_new.xlsx로 저장
    covid19_1.to_excel(writer, sheet_name= 'sheet1', index=False) # 원본 데이터는 sheet1에 저장
    covid19_2.to_excel(writer, sheet_name= 'sheet2', index=False) # 중국 데이터는 sheet2에 저장
    covid19_3.to_excel(writer, sheet_name= 'sheet3', index=False) # 타국 데이터는 sheet3에 저장
    covid19_4.to_excel(writer, sheet_name= 'sheet4', index=False) # 한국 데이터는 sheet4에 저장
```

```python
## auto-mpg.csv 파일은 자동차의 연비에 대한 데이터 파일이다.

import pandas as pd

# 다음에 맞추어서 해당 데이터셋에 columns name을 설정하고, auto-mpg-2.csv로 저장하기
auto_mpg = pd.read_csv('C:\\Users\\82107\\Desktop\\5주차QUEST_데이터셋\\auto-mpg.csv') # auto-mpg.csv 읽어옴
column_name = ['mpg(연비)', 'cylinders(실린더수)', 'displacement(배기량)', 'horsepower(출력)', 'weight(차중)', 'acceleration(가속능력)', 'modelYear(출시년도)', 'originNumber(제조국번호)', 'name(모델명)'] # 컬럼 이름 설정
data = auto_mpg.values.tolist() # 데이터 설정
auto_mpg = pd.DataFrame(data, columns=column_name) # auto_mpg에 데이터와 컬럼 이름 재설정 

auto_mpg.to_csv("C:\\Users\\82107\\Desktop\\auto-mpg-2.csv", encoding="utf-8-sig", index=False) # auto-mpg-2.csv로 저장

# 80년대에 생산된 모델의 모델명만을 출력하라
auto_mpg2 = pd.read_csv('C:\\Users\\82107\\Desktop\\auto-mpg-2.csv') # auto-mpg-2.csv 읽어옴
is_80years = auto_mpg2['modelYear(출시년도)'] >= 80 
is_80years_ok = auto_mpg2[is_80years] # 80년대에 생산된 모델 가져옴
print('-------------------80년대에 생산된 모델의 모델명-------------------')
print(is_80years_ok[['name(모델명)']], end="\n\n\n\n\n") # 모델명만을 출력

# mpg(연비)와 horsepower(출력) 열의 표준편차값을 구하라
print('------------mpg(연비)와 horsepower(출력) 열의 표준편차값------------')
print(auto_mpg2[['mpg(연비)']].std()) # 연비의 표준편차값 출력
print(auto_mpg2[['horsepower(출력)']].std(), end="\n\n\n\n\n") # 출력의 표준편차값 출력

# horsepower(출력) 열의 표준편차는 구할 수 없다. 이유를 서술하라.
print('-----------horsepower(출력) 열의 표준편차는 구할 수 없는 이유------------')
print("물음표값이 포함되어 있었음")

# modelYear(출시년도)를 최신순에 따라 데이터를 정렬해서 auto-mpg-3.csv로 저장하라
auto_mpg2 = auto_mpg2.sort_values('modelYear(출시년도)', ascending=False) # 출시년도에 따라 정렬
auto_mpg2.to_csv("C:\\Users\\82107\\Desktop\\auto-mpg-3.csv", encoding="utf-8-sig", index=False) # auto-mpg-3.csv로 저장
```

```python
## 남북한발전전력량.xlsx 파일엔 남한과 북한의 발전전력별 연간통계량이 저장되어있다.

import pandas as pd

# 해당 데이터를 읽어와서 남한과 북한의 수력 및 화력발전과 그 합계에 대한 연간 통계량만을 분리해서 남북한수력및화력발전전력량.csv로 저장하라.
korea = pd.read_excel('C:\\Users\\82107\\Desktop\\5주차QUEST_데이터셋\\남북한발전전력량.xlsx') # 남북한발전전력량.xlsx를 읽어옴

korea_energe = korea.loc[[1,2]] # 남한의 수력, 화력 데이터
south = korea.loc[1] + korea.loc[2] # 남한의 수력, 화력 합계 데이터 계산
korea_energe = korea_energe.append(south, ignore_index=True) # 남한의 수력, 화력 합계 데이터 추가

korea_energe = korea_energe.append(korea.loc[[6,7]], ignore_index=True) # 북한의 수력, 화력 데이터 추가
north = korea.loc[6] + korea.loc[7] # 북한의 수력, 화력 합계 데이터 계산
korea_energe = korea_energe.append(north, ignore_index=True) # 남한의 수력, 화력 합계 데이터 추가

korea_energe = korea_energe.reindex(index = [2,0,1,5,3,4]) # 합계가 먼저 오도록 순서 정렬

# 엑셀을 확인하기 쉽도록 라벨링 작업
korea_energe.iloc[0]['전력량 (억㎾h)'] = '남한'
korea_energe.iloc[0]['발전 전력별'] = '합계'
korea_energe.iloc[3]['전력량 (억㎾h)'] = '북한'
korea_energe.iloc[3]['발전 전력별'] = '합계'

korea_energe.to_csv("C:\\Users\\82107\\Desktop\\남북한수력및화력발전전력량.csv", encoding="utf-8-sig", index=False) # 남북한수력및화력발전전력량.csv로 저장
```

```python
## 다음을 만족하는 20년 달력 데이터 프레임을 만드시오.

import pandas as pd
import calendar
import datetime

# 계절 Validation
def season(monthNum) :
    if monthNum == 12 or monthNum == 1 or monthNum == 2:
        return 'winter'
    elif monthNum == 3 or monthNum == 4 or monthNum == 5:
        return 'spring'
    elif monthNum == 6 or monthNum == 7 or monthNum == 8:
        return 'summer'
    elif monthNum == 9 or monthNum == 10 or monthNum == 11:
        return 'fall'

# 주말 Validation
def weekend(weekday) :
    if weekday == 'Sat' or weekday == 'Sun' :
        return 1
    else :
        return 0

cal = pd.DataFrame(columns=['original', 'Year', 'Month', 'Day', 'weekday', 'weekend', 'season']) # 컬럼명 설정
for j in range(1, 13): # 1월부터 12월까지 반복
    for i in range(1, calendar.monthrange(2020, j)[1] + 1): # 각 달의 최종일수까지 반복
        date = datetime.date(2020, j, i) # 순차적으로 일수 저장
        cal = cal.append(pd.DataFrame([[date.strftime("%y-%m-%d"), date.strftime("%y"), date.month, date.day, date.strftime("%a"), weekend(date.strftime("%a")), season(date.month)]], columns=['original', 'Year', 'Month', 'Day', 'weekday', 'weekend', 'season']), ignore_index=True) # 행 추가

cal # 확인

# cal.to_csv("C:\\Users\\82107\\Desktop\\2020달력.csv", encoding="utf-8-sig", index=False)
```
